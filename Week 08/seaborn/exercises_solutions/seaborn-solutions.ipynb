# Exercises

## 1.

Show the distribution of the points given by the reviewers in the dataset.

```python
# Show the distribution of the points given by the reviewers in the dataset.
plt.figure(figsize=(10, 6))

plt.title("Distribution of Wine Review Scores")

sns.histplot(reviews.points, bins=20, kde=False)
```

## 2.

For each reviewer in the dataset show how many reviews they have written.

```python
# For each reviewer in the dataset show how many reviews they have written.

plt.figure(figsize=(10, 6))

plt.title("Number of Reviews per Reviewer")

sns.barplot(y=reviews.taster_name.value_counts().index, x=reviews.taster_name.value_counts())
```

## 3.

Using scatterplot show the relationship between the price and points for each wine.

```python
# 
plt.figure(figsize=(10, 6))

plt.title("Price vs Points")

sns.scatterplot(x=reviews.price, y=reviews.points)
```
